## Fetching population data from Wikipedia

In this example we will fetch data about countries and their population from Wikipedia.

http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population has several tables for individual countries, subcontinents as well as different years. We will combine the data for all countries and all years in a single panda dataframe and visualize the change in population for different countries.

###We will go through the following steps:
* fetching html with embedded data
* parsing html to extract the data
* collecting the data in a panda dataframe
* displaying the data

To give you some starting points for your homework, we will also show the different sub-steps that can be taken to reach the presented solution.

## Fetching the Wikipedia site

```python
#set url, use requests to get url into text format
url = 'http://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population'
website_html = requests.get(url).text
#print website_html
```

## Parsing html data

```python
# in order to parse the data you need to take each element and set them into dom's you can iterate over
def get_population_html_tables(html):
    """Parse html and return html tables of wikipedia population data."""

    dom = web.Element(html)

    ### 0. step: look at html source!
    # do a search on "<table"
    # or you can right click on browser and use inspect elemnt
    #### 1. step: get all tables
    ##tbls=dom('table')
    #previous command uses pattern library to find element/tag = table
    #this is important to find out the tables you may care for
    tbls=dom.by_class('sortable wikitable')
    # previous command uses pattern library to find element by class = sortable wikitable
    #tbls=dom.by_class('sortable wikitable jquery-tablesorter')

    #### 2. step: get all tables we care about

    return tbls

tables = get_population_html_tables(website_html)
print "table length: %d" %len(tables)
for t in tables:
    print t.attributes

    
```

```python
def table_type(tbl):
    ### Extract the table type
    #return the second elements comment with tag 'th'
    #returns second table and its content which are the columns
    #returns the every tables first table header's content(text output)
    return tbl('th')[0].content




# group the tables by type
#default dicts is good for making new dictionaries/lists
#dict table type, multiple tables, check if key in dictionary
#if not there create it, if there then append to current key
tables_by_type = defaultdict(list)  # defaultdicts have a default value that is inserted when a new key is accessed
#if there is new element in the dict, it creates an empty list
for tbl in tables:
    tables_by_type[table_type(tbl)].append(tbl)
    #defaultdict[sets key].appends(content)

print tables_by_type
#prints out multiple entries for each of the column header
#note there country or territory corresponds to all the atable headers that are
#country or territory. C&T has 3 tables
#print tables_by_type['Country or territory'][0]
#prints out country or territories first instance
```

## Extracting data and filling it into a dictionary

```python
def get_countries_population(tables):
    """Extract population data for countries from all tables and store it in dictionary."""
    
    result = defaultdict(dict)
         
        ## 1. step: try to extract data for a single table
        #
        #tbl=tables[0]
        ##call first table
    for tbl in tables:   
        headers=tbl('tr')
        #call all rows
        first_header=headers[0]
        #call first row
        th_s=first_header('th')
        #call the th's of first row
        years=[int(val.content) for val in th_s if  val.content.isnumeric()]
        year_indices=[idx for idx,val in enumerate(th_s)if val.content.isnumeric()]
    
        ''' 
        enumerate returns (index,value) as a tuple
        read above with the first part as the append then the rest forward nesting
        for val in th_s:
            if val.content.is_isnumeric():
                hidden_list.append(int(val.content))
        '''
        #table_headers=tbl('thead')[0]('th')
        '''
        tbl is dom element
        headers - searches through dom for children 'thead'
            (returns list). does not matter for nesting. Beautiful soup
        first_headers - returns the first one (index 0)
        head
        '''
        #years=[int(th.content)]
        print years
        print year_indices
        #note: getting first table header in first table
        # 2. step: iterate over all tables, extract headings and actual data and combine data into single dict
        rows=tbl('tr')[1:]
        for row in rows:
            tds = row('td')
            country_name=tds[0]('a')[0].content
            population_by_year=[int(tds[colidx].content.replace(',','')) for colidx in year_indices]
            #pull population for only year indices
            subdict=dict(zip(years,population_by_year))
            #zip -creates nesting dictionary
            result[country_name].update(subdict)
    return result


result = get_countries_population(tables_by_type['Country or territory'])
#print result
print len(result),'Countries extracted'
print result[u'Canada']
```

## Plotting population of 4 countries

```python
plotCountries = ['Austria', 'Germany', 'United States', 'France']
    
for country in plotCountries:
    row = df.ix[country]
    plt.plot(row.index, row, label=row.name ) 
    
plt.ylim(ymin=0) # start y axis at 0

plt.xticks(rotation=70)
plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("# people (million)")
plt.title("Population of countries")
```

## Plot 5 most populous countries from 2010 and 2060

```python
def plot_populous(df, year):
    # sort table depending on data value in year column
    df_by_year = df.sort(year, ascending=False)
    #sort dataframe by the values in 'year'
    plt.figure()
    for i in range(5):  
        row = df_by_year.ix[i]
        plt.plot(row.index, row, label=row.name ) 
            
    plt.ylim(ymin=0)
    
    plt.xticks(rotation=70)
    plt.legend(loc='best')
    plt.xlabel("Year")
    plt.ylabel("# people (million)")
    plt.title("Most populous countries in %d" % year)

plot_populous(df, 2010)
plot_populous(df, 2050)
```
